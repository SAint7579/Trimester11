```python
#Importing the function
import numpy as np
import matplotlib.pyplot as plt
```

## Triangle

```python
##Triangle MF
def TriangleMF(x,a,b,c):
    return max(min([(x-a)/(b-a) , (c-x)/(c-b)]),0)

#Plot Triangle Graph
def plotTriangle(a,b,c):
    value = []
    for x in range(1,100):
        value.append(TriangleMF(x,a,b,c))
    
    #Ploting the output
    plt.plot(range(1,100),value)
    plt.show()
```

```python
#Testing the function
plotTriangle(30,55,80)
plotTriangle(0,50,100)
plotTriangle(9,10,100)
```

## Trapezoidal

```python
##Trapezoid MF
def TrapezoidMF(x,a,b,c,d):
    return max(min([(x-a)/(b-a) , 1 , (d-x)/(d-c)]),0)

#Plot Trapezoid Graph
def plotTrapezoid(a,b,c,d):
    value = []
    for x in range(1,100):
        value.append(TrapezoidMF(x,a,b,c,d))
    
    #Ploting the output
    plt.plot(range(1,100),value)
    plt.show()
```

```python
#Testing the function
plotTrapezoid(30,55,80,90)
plotTrapezoid(10,11,89,90)
```

## Gaussian

```python
##Gaussian MF
def GaussianMF(x,c,sig):
    return np.exp(-0.5 * (((x-c)/sig)**2))

#Plot Gaussian Graph
def plotGaussian(c,sig):
    value = []
    for x in range(1,100):
        value.append(GaussianMF(x,c,sig))
    
    #Ploting the output
    plt.plot(range(1,100),value)
    plt.show()
```

```python
#Testing the function
plotGaussian(40,10)
plotGaussian(40,30)
plotGaussian(70,30)
```

## Sigmoid

```python
##SigmoidMF
def SigmoidMF(x,a,c):
    return 1 / (1 + np.exp(-a*(x-c)))

#Plot Sigmoid Graph
def plotSigmoid(a,c):
    value = []
    for x in np.arange(1,10,0.01):
        value.append(SigmoidMF(x,a,c))
    
    #Ploting the output
    plt.plot(np.arange(1,10,0.01),value)
    plt.show()
```

```python
#Testing the function
plotSigmoid(2,5)
plotSigmoid(1,5)
plotSigmoid(-4,5)
```
